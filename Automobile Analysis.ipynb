```python
import numpy as np
import pandas as pd
```

```python
# import Data file in python with the help of pandas
Data=pd.read_csv("Automobile_data.csv")
DATA=pd.DataFrame(Data)
DATA.info()
DATA["price"]=DATA["price"].fillna(0)
DATA.info()

```

```python
# average mileage and average price of convertible cars
avg_Price=DATA.groupby("body-style")["price","average-mileage"].mean()
average_Price_mileage=avg_Price.head(1)
average_Price_Mileage=pd.DataFrame(average_Price_mileage)
average_Price_Mileage
```

```python
# Sort cars by company and price in descending order
cars=DATA.sort_values("price",ascending=False)
cars
```

```python
# Save data for 'Toyota' separately...
Toyota=DATA[DATA["company"]=="toyota"]
Toyota
```

```python
# most expensive car
Cars=cars.head(1)
Cars
```

```python
# the cars on basis of maximum length, highest wheelbase, maximum horsepower
ml_hw_hp=DATA.groupby(["company","body-style"])["wheel-base","length","horsepower"].max()
ml_hw_hp
```

```python
#company wise average price of cars and order them in decending order
avg_price=DATA.groupby("company")["price"].mean()
Avg_price=pd.DataFrame(avg_price)
Average_price=Avg_price.sort_values("price",ascending=False)
Average_price
```

```python
#7. create a column named: "decision", If wheel-base is less than 100 and length is less than 175 then "don't buy" else "buy"
DATA.loc[(DATA["wheel-base"]<100) & (DATA["length"]<175),"decision"]="Dont buy"
DATA.loc[(DATA["wheel-base"]>100) & (DATA["length"]>175),"decision"]="buy"
DATA
```

```python
# Select only cars whose decision is "buy"; find average price based on company of car..
DATA
buy=DATA[DATA["decision"]=="buy"]
average_price_of_buy_cars=buy.groupby("company")["price"].mean()
average_price=pd.DataFrame(average_price_of_buy_cars)
average_price
```
